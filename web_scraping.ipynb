```python
data_list = []
final = pd.DataFrame()
for j in range(1, 6):
    webpage = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={j}', headers=header).text
    soup = BeautifulSoup(webpage, 'html.parser')
    company = soup.find_all('div', class_="companyCardWrapper")
    
    name = []
    rating = []
    rating_count = []
    job_type = []
    loc = []
    salary = []
    job_no = []
    
    for i in company:
        
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_="rating_star_container").text.strip())
        rating_count.append(i.find('span', class_="companyCardWrapper__companyRatingCount").text.strip())
        job_type.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip().split('|')[0])
        loc.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip().split('|')[1])
        salary.append(i.find_all('a', class_="companyCardWrapper__ActionWrapper")[1].text)
        try:
            job_no.append(i.find_all('a', class_="companyCardWrapper__ActionWrapper")[3].text)
        except:
            job_no.append(np.nan)

    df = pd.DataFrame({'Company Name': name,
    'Rating': rating,
    'Rating Count': rating_count,
    'Type': type,
    'Location': loc,
    'Median Salaries': salary,
    'No of jobs': job_no})

    data_list.append(df)
    
final = pd.concat(data_list, ignore_index= True)
```
